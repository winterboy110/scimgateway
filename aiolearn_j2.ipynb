```python
import matplotlib.pyplot as plt
import cv2
#image = cv2.imread('image2.jpg', cv2.IMREAD_GRAYSCALE)
image = cv2.imread('image2.jpg', cv2.COLOR_BGR2RGB)

plt.imshow(image)

model = cv2.CascadeClassifier('aiolearn.xml')
face = model.detectMultiScale(image)
print(face)
x = face[0][0]
y = face[0][1]
a = face[0][2]
b = face[0][3]
image = cv2.rectangle(image, (x, y), (x + a, y + b), (100, 100, 100), 2)
plt.imshow(image)
```
